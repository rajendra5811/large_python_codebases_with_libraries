Question 1: Vehicle Rental Management System
Interview question

Design a vehicle rental system for a company that rents cars and electric cars.

Create a base class called Vehicle with these attributes:

vehicle_id
brand
model
year
daily_rate
is_available
mileage
rental_count

The class should provide methods to:

Rent the vehicle for a given number of days.
Return the vehicle and update its mileage.
Calculate the rental cost.
Display vehicle details.
Prevent an unavailable vehicle from being rented.
Prevent invalid rental days or mileage values.

Create a subclass called ElectricVehicle with these additional attributes:

battery_capacity
battery_percentage

Override the rental-cost calculation so that electric vehicles receive a 10% environmental discount.

Also create a Customer class that stores customer information and rented vehicles.

The customer should not be allowed to rent more than two vehicles simultaneously.

Expected concepts

This question tests:

Classes and objects
Encapsulation
Inheritance
Method overriding
Polymorphism
Composition
Class variables
Class methods
Static methods
Custom exceptions
